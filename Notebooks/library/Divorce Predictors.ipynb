# 3. Data Analysis and Visualization

We'll perform deeper analysis to explore these questions, utilizing visualizations to illustrate trends, correlations, and other insights.

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Calculate correlations
correlation_matrix = divorce_data.corr()

# Plot heatmap of the correlation matrix focusing on the 'Class' column
plt.figure(figsize=(8, 12))
heatmap = sns.heatmap(correlation_matrix[['Class']].sort_values(by='Class', ascending=False), 
                      vmin=-1, vmax=1, annot=True, cmap='coolwarm')
heatmap.set_title('Correlation with Marital Status', fontdict={'fontsize':18}, pad=16);
plt.show()
```

The heatmap displays the correlations between each survey attribute and the marital status ('Class'). Attributes with stronger positive correlations suggest factors that might be associated with being married, while negative correlations suggest factors associated with being divorced.

### Key Observations: 

Several attributes show strong correlations (both positive and negative) with marital status. This suggests that these factors are particularly influential in the dynamics of marriage and its success or dissolution.
The stronger positive correlations indicate attributes where higher scores (agreement or presence of positive behaviors) are linked with being married.
Conversely, stronger negative correlations might represent attributes where higher scores (possibly reflecting negative behaviors or disagreements) are associated with being divorced.

Let's create visualizations for the top positively and negatively correlated attributes to better understand their impact on marital status. 

```python
# Select top 3 positively and negatively correlated attributes
top_correlated = correlation_matrix['Class'].sort_values(key=abs, ascending=False).head(4).index.tolist()

# Plot distribution plots for these attributes
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(14, 10))
fig.suptitle('Distribution of Responses for Top Correlated Attributes', fontsize=16)

for i, atr in enumerate(top_correlated):
    sns.histplot(divorce_data, x=atr, hue="Class", element="step", stat="density", common_norm=False, ax=axes[i//2, i%2])
    axes[i//2, i%2].set_title(f'Distribution of {atr}')
    axes[i//2, i%2].set_xlabel('Response Score')
    axes[i//2, i%2].set_ylabel('Density')

plt.tight_layout(rect=[0, 0.03, 1, 0.95])
plt.show()
```

### Analysis of the Histograms: 

- Attributes with high positive correlations to being married show a trend where higher scores (indicating stronger agreement or presence of positive behaviors) are more common among those who are married.
- Conversely, attributes with high negative correlations to being married show that higher scores (potentially indicating negative behaviors or conflicts) are more common among those who are divorced.
